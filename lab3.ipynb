```python

import pandas as pd
data = pd.read_csv('C:/Users/Xena/LabsPython/.ipynb_checkpoints/NationalNames.csv')

# Групуємо дані за роками та статтю і підсумовуємо кількість
births_per_year_gender = data.groupby(['Year', 'Gender'])['Count'].sum().unstack()

# Підраховуємо кількість років, коли дівчаток народжувалось більше, ніж хлопчиків
years_more_female_births = (births_per_year_gender['F'] > births_per_year_gender['M']).sum()

# Виводимо результат
print(years_more_female_births)
```

```python

import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('C:/Users/Xena/LabsPython/.ipynb_checkpoints/NationalNames.csv')

# Групуємо дані за роками та статтю і підсумовуємо кількість
births_per_year_gender = data.groupby(['Year', 'Gender'])['Count'].sum().unstack()

# Створюємо графік
births_per_year_gender.plot()

# Додаємо заголовки та підписи осей
plt.title('Total births per year by gender')
plt.xlabel('Year')
plt.ylabel('Total births per year')

# Виводимо графік
plt.show()
```

```python

import pandas as pd

data = pd.read_csv('C:/Users/Xena/LabsPython/.ipynb_checkpoints/NationalNames.csv')

# Знаходимо імена, які використовувалися для обох статей
gender_neutral_names = data.groupby('Name')['Gender'].nunique()

# Відбираємо лише ті імена, які використовувалися для обох статей
neutral_names = gender_neutral_names[gender_neutral_names == 2].index

# Підраховуємо кількість років для кожного гендерно-нейтрального імені
names_year_count = data[data['Name'].isin(neutral_names)].groupby('Name')['Year'].nunique()

# Відбираємо імена, які використовувалися в усі роки для обох статей
# (фільтруємо ті, що присутні протягом усього періоду, якщо це потрібно)
most_popular_neutral_names = data[data['Name'].isin(neutral_names)].groupby('Name')['Count'].sum().sort_values(ascending=False)

# Виводимо список найпопулярніших гендерно-нейтральних імен
print(most_popular_neutral_names.head(15).index)
```

```python

import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('C:/Users/Xena/LabsPython/.ipynb_checkpoints/NationalNames.csv')

# Фільтруємо дані для імен John та Mary
john_data = data[data['Name'] == 'John'].groupby('Year')['Count'].sum()
mary_data = data[data['Name'] == 'Mary'].groupby('Year')['Count'].sum()

# Побудова графіка
plt.plot(john_data.index, john_data.values, label='John')
plt.plot(mary_data.index, mary_data.values, label='Mary')

# Додаємо заголовки та підписи
plt.title("Distribution of the names 'John' and 'Mary' over the years")
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend(title='Name')

# Відображаємо графік
plt.show()
```
