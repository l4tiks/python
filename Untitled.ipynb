```python
import csv
import numpy as np

def load_column(filename, column_name):
    """
    Зчитує стовпець з CSV-файлу за назвою.
    
    Parameters:
        filename (str): Назва файлу CSV.
        column_name (str): Назва стовпця, який потрібно зчитати.
        
    Returns:
        np.ndarray: Масив numpy зі значеннями стовпця.
    """
    with open(filename, 'r', encoding='utf-8') as file:
        reader = csv.reader(file)
        headers = next(reader)  # Пропускаємо заголовки
        try:
            col_index = headers.index(column_name)
        except ValueError:
            raise ValueError(f"Column '{column_name}' not found in file")
        
        data = [int(row[col_index]) for row in reader if row[col_index].isdigit()]
    return np.array(data)
```

```python
filename = "russia_losses_equipment.csv"
apc_losses = load_column(filename, "APC")
print(apc_losses[:10])  # Перевірка: перші 10 значень
```

```python
def daily_losses(losses):
    """
    Обчислює добові втрати на основі загальних даних.
    
    Parameters:
        losses (np.ndarray): Массив із загальними втратами.
        
    Returns:
        np.ndarray: Масив добових втрат.
    """
    return np.diff(losses, prepend=0)

daily_apc_losses = daily_losses(apc_losses)
print(daily_apc_losses[:10])  # Перевірка: перші 10 значень
```

```python
def top_losses(losses, n=5):
    """
    Знаходить найбільші значення у масиві втрат.
    
    Parameters:
        losses (np.ndarray): Масив добових втрат.
        n (int): Кількість найбільших втрат, які потрібно знайти.
        
    Returns:
        np.ndarray: Масив із n найбільших значень.
    """
    return np.sort(losses)[-n:][::-1]  # Сортування та зворотній порядок

top_5_losses = top_losses(daily_apc_losses)
print("5 найбільших добових втрат БТР:", top_5_losses)
```

```python
import pandas as pd

def summer_losses_2023(filename, losses_column):
    """
    Рахує втрати БТР за літо 2023 року.
    
    Parameters:
        filename (str): Назва файлу CSV.
        losses_column (str): Назва стовпця з втратами.
        
    Returns:
        int: Загальні втрати за літо 2023 року.
    """
    data = pd.read_csv(filename, parse_dates=['date'])
    summer_data = data[(data['date'] >= "2023-06-01") & (data['date'] <= "2023-08-31")]
    return summer_data[losses_column].diff().fillna(0).sum()

summer_losses = summer_losses_2023(filename, "APC")
print("Втрати БТР влітку 2023 року:", int(summer_losses))
```

```python
def average_losses(losses, start_day, end_day):
    """
    Обчислює середнє значення втрат у заданому діапазоні днів.
    
    Parameters:
        losses (np.ndarray): Масив загальних втрат.
        start_day (int): Початковий день (індекс).
        end_day (int): Кінцевий день (індекс).
        
    Returns:
        float: Середнє значення втрат.
    """
    return np.mean(losses[start_day:end_day])

avg_losses = average_losses(daily_apc_losses, 100, 500)
print("Середнє значення втрат за 100-500 днів:", avg_losses)
```

```python
import matplotlib.pyplot as plt

def plot_losses(losses, last_days=200):
    """
    Створює графік втрат за останні N днів.
    
    Parameters:
        losses (np.ndarray): Масив добових втрат.
        last_days (int): Кількість днів для візуалізації.
    """
    plt.figure(figsize=(8, 16), dpi=100)
    plt.plot(losses[-last_days:], linestyle='--', color='blue', label="Втрати БТР")
    
    # Налаштування графіку
    plt.title("Втрати БТР за останні 200 днів")
    plt.xlabel("Дні")
    plt.ylabel("Кількість втрат")
    plt.grid(color='gray', linestyle='-', linewidth=0.5)
    plt.legend()
    plt.savefig("apc_losses_last_200_days.png")
    plt.show()

plot_losses(daily_apc_losses)
```
